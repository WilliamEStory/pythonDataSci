```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5, 11)
```

```python
y = x ** 2
```

```python
x
```

```python
y
```

```python
# functional method for generating plots
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# Subplots
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r-')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# Object Oriented Method
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```
